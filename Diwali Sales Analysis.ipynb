### Exploratory Data Analysis

#### Gender

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

##### From the above graph, we can see that females have made more purchases than males

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

##### From this graph, we can see that women also have more purchasing power than men

### Age

```python
ax = sns.countplot(x = 'Age Group', data = df)
```


##### The graph shows maximum number of purchases are made by the people belonging to the age group 26-35

```python
ax = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

##### From the above graph we can see that most of the buyers are from the age group of 26-35 (who are mostly women)

### State

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
```

##### From the graph, we see the top 10 states who have placed the most number of orders. The most orders are placed by Uttar Pradesh and second most are from Maharashtra.

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'State', y = 'Amount', data = sales_state)
```

##### Despite Kerala placing more orders than Haryana, Bihar and Gujrat, it does not appear in this graph. This means that Haryana, Bihar and Gujrat have more purchasing power than Kerala.

### Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(2,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', data = sales_state, hue = 'Gender')
```

##### The graph shows that most of the shopping was done by married women

### Occupation

```python
ax = sns.countplot(data = df, x = 'Occupation')

sns.set(rc={'figure.figsize':(23,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

##### The most number of orders were placed by the people working in IT Sector.

```python

ax = sns.countplot(data = df, x = 'Occupation', hue = 'Gender')

sns.set(rc={'figure.figsize':(23,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

##### This is a more detailed graph which shows the occupation as well as the gender of the buyers. We can see that most buyers were women working in IT Sector

```python
sales_state = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Occupation', y = 'Amount', data = sales_state)
```

##### People working in IT Sector hold the maximum purchasing power, followed by people working in healthcare.

### Product Category

```python
ax = sns.countplot(data = df, x = 'Product_Category')

sns.set(rc={'figure.figsize':(23,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_Category', y = 'Amount', data = sales_state)
```

```python
sales_state = df.groupby(['Product_ID'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'Product_ID', y = 'Amount', data = sales_state)
```

### Conclusion

##### Married women of the age 26-35 belonging to the states Uttar Pradesh, Maharashtra and Karnataka working in IT Sector, Healthcare and Aviation are more likely to buy products from food, clothing and electronics category.
